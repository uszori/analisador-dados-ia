```python
import pandas as pd

df = pd.read_csv(
    "../data/MICRODADOS_ENEM_2023.csv",
    encoding="latin-1",
    sep=";",
    nrows=10000  # carrega só 10 mil linhas pra não travar
)

df.head()
```

```python
colunas_notas = ['NU_NOTA_CN', 'NU_NOTA_CH', 'NU_NOTA_LC', 'NU_NOTA_MT', 'NU_NOTA_REDACAO']
df[colunas_notas].describe()
```

```python
import matplotlib.pyplot as plt

df[colunas_notas].hist(bins=30, figsize=(14, 8), color='steelblue', edgecolor='white')
plt.suptitle('Distribuição das Notas - ENEM 2023', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='TP_ESCOLA', y='NU_NOTA_MT')
plt.title('Nota de Matemática por Tipo de Escola - ENEM 2023')
plt.xlabel('Tipo de Escola (1=Federal, 2=Estadual, 3=Municipal, 4=Privada)')
plt.ylabel('Nota Matemática')
plt.show()
```

```python
colunas_notas = ['NU_NOTA_CN', 'NU_NOTA_CH', 'NU_NOTA_LC', 'NU_NOTA_MT', 'NU_NOTA_REDACAO']
df[colunas_notas].isnull().sum()
```

```python
df_limpo = df.dropna(subset=colunas_notas)
df_limpo = df_limpo[df_limpo[colunas_notas].min(axis=1) > 0]

print(f"Antes: {len(df)} candidatos")
print(f"Depois: {len(df_limpo)} candidatos")
```

```python
media_por_estado = df_limpo.groupby('SG_UF_PROVA')['NU_NOTA_MT'].mean().sort_values(ascending=False)

print(media_por_estado)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
media_por_estado.plot(kind='bar', color='steelblue', edgecolor='white')
plt.title('Média de Matemática por Estado - ENEM 2023')
plt.xlabel('Estado')
plt.ylabel('Média')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
